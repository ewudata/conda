```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

donations = pd.read_csv("wa.csv",  low_memory=False,  index_col=False)
donations.shape
donations.head(2)
total_donations = donations.groupby("cand_nm").sum().sort_values("contb_receipt_amt")
print(total_donations)
total_donations["contb_receipt_amt"].plot(kind="bar")
```

```python
avg_donations = donations.groupby("cand_nm").mean().sort_values(by="contb_receipt_amt")
avg_donations["contb_receipt_amt"].plot(kind="bar")
```
